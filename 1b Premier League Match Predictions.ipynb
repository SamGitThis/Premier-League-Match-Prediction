```python
matches['Date'] = pd.to_datetime(matches['Date'])
```

```python
matches['venue_code'] = matches['Venue'].astype('category')
```

```python
matches['venue_code'] = matches['venue_code'].cat.codes
```

```python
matches['opp_code'] = matches['Opponent'].astype('category').cat.codes
```

```python
matches['hour'] = matches['Time'].str.replace(":.+", "", regex = True).astype('int')
#":.+" the .+ tells the the system to replace one or more chracters after the :
#":." will only replace the first element after :
```

```python
matches.dtypes
```

```python
matches['day_code'] = matches['Date'].dt
```

```python
matches.head()
```

```python
matches['day_code'] = matches['Date'].dt.dayofweek
```

```python
matches.head()
```

```python
matches['target'] = (matches['Result'] == "W").astype('int')
```

```python
matches
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf = RandomForestClassifier(n_estimators = 50, min_samples_split = 10, random_state = 1)
```

```python
train = matches[matches["Date"] < '2024-01-01']
```

```python
test = matches[matches['Date'] > '2024-01-01']
```

```python
predictors = ['venue_code', 'opp_code', 'hour', 'day_code']
```

```python
rf.fit(train[predictors], train['target'])
```

```python
preds = rf.predict(test[predictors])
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_score
```

```python
accuracy = accuracy_score(test['target'], preds)
```
